```
import numpy as np
import pandas as pd 
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing 
```

```
df=pd.read_csv('/content/FertilizerPrediction.csv')
df1=pd.read_csv('/content/MergeFileCrop.csv')
```

```
df1.columns
```

```
le = preprocessing.LabelEncoder()
df['Crop']=le.fit_transform(df['Crop Type'])
df['Fertilizer']=le.fit_transform(df['Fertilizer Name'])
df1['crop']=le.fit_transform(df1['label'])
```

```
X=df[['Temparature', 'Humidity ']]
Y= df['Fertilizer']
```

```
lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
```

```
test=df1[['temperature','humidity']]
```

```
model.predict(test)
```
