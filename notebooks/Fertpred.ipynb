```
import numpy as np
import pandas as pd 
from sklearn import linear_model
from sklearn import preprocessing 
from sklearn.model_selection import train_test_split
```

```
df=pd.read_csv('../data/Fertilizer Prediction.csv')
```

```
df.head()
```

```
df.columns
```

```
le1 = preprocessing.LabelEncoder()
le2 = preprocessing.LabelEncoder()
# df['Crop']=le.fit_transform(df['Crop Type'])
df['Fertilizer']=le1.fit_transform(df['Fertilizer Name'])
df['Soil']=le1.fit_transform(df['Soil Type'])
```

```
features=df[['Nitrogen','Potassium','Phosphorous','Soil','Temparature','Humidity ']]
target=df['Fertilizer']
```

```
features.isna().any()
```

```
from sklearn.preprocessing import StandardScaler
import pickle
scalerfile = 'scalerknn.sav'
scaler = StandardScaler()
scaler.fit(features)
pickle.dump(scaler, open(scalerfile, 'wb'))
scaled=scaler.transform(features)
X_train, X_test, y_train, y_test = train_test_split(
    scaled, target, test_size=0.1, random_state=13)
```

```
from sklearn.neighbors import KNeighborsClassifier
# lm = linear_model.LinearRegression()
lm=KNeighborsClassifier(n_neighbors=4)
model = lm.fit(X_train,y_train)
```

```
preds=model.predict(X_test)
preds
```
